```python
EPS = 1e-5
a = 2
b = 3
 
```

```python
def newton(f, a, b, EPS):
    x1 = a
    x2 = (a + b) / 2
    df = f.derivative()
    while abs(x2 - x1) >= EPS:
        x1 = x2
        x2 = x1 - f(x=x1)/df(x=x1)

    return x1
```

```python
 
x_1 = newton(func, a, b, EPS)
show(x_1.n())
show("f=",func)
p1 = plot(func,(-5,7))
p1 += point((x_1, 0), color="red", size=30, zorder=20)
p1 += text(r"$x_1$", (x_1, -0.1), color="black", fontsize=14)

p1.show()
```

```python
eq1 = 1/abs(func.diff())
y = plot(eq1, (2, 3), ymin=6, ymax=8)
A=eq1(x = 2)+EPS
a=plot(A,(2, 3), ymin=6, ymax=8,color="red",)
show(y+a)
show("A=", A.n())
 
```

$1)\ \frac{1}{|f'(x)|} < A =>   -\ существует\ f'(x) \ \ и\ не\ равна\ нулю$

```python
diff_1 = func.diff()
s1 = func/diff_1
eq2 = abs(s1)
d1 = plot(eq2, (2, 3), ymin=0, ymax=8, color ="blue")
B=eq2(x = 2)+0.00001
b=plot(B,(2, 3), ymin=0, ymax=8,color="red",)
show(d1+b)
show("B=", B.n())
```

$2)\ \left|\frac{f(x)}{f'(x)}\right| < B =>\ f(x)\ -\ ограничена$

```python
diff_2 = diff_1.diff()
d2 = plot(diff_2, (2, 3), ymin=-1, ymax=1, color="blue")
C=diff_2(x = 3)+0.00001
D = 1/(2*A*B)
d= plot(D, (2, 3), ymin=-1, ymax=1,  color="red")
c= plot(C, (2, 3), ymin=-1, ymax=1,  color="green")
show(d2+c+d)
show("C =", C.n())
show("1/2AB =", D.n())
```
